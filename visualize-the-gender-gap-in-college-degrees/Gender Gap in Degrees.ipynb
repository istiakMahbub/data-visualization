## üéì Gender Distribution in College Majors (1970‚Äì2010)

This notebook analyzes the gender distribution in various **college majors** from 1970 to 2010. The majors are grouped into three main categories:

### üìÅ Category Groups

- #### üß† **STEM Categories (`stem_cats`)**
  - Psychology, Biology, Math and Statistics, Physical Sciences, Computer Science, Engineering

- #### üé® **Liberal Arts Categories (`lib_arts_cats`)**
  - Foreign Languages, English, Communications and Journalism, Art and Performance, Social Sciences and History

- #### üèõÔ∏è **Other Categories (`other_cats`)**
  - Health Professions, Public Administration, Education, Agriculture, Business, Architecture

Each major shows the percentage of **women graduates**, and we calculate the percentage of **men graduates** as `100 - women_percentage`.

```python
import pandas as pd
data = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
data.head()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, len(stem_cats), figsize=(25, 5))

for i, cat in enumerate(stem_cats):
    x = data["Year"]
    y = data[cat]
    z = 100 - y
    axes[i].plot(x, y, label="Women")
    axes[i].plot(x, z, label="Men")
    axes[i].set_title(cat)
    axes[i].legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, len(lib_arts_cats), figsize=(25, 5))

for i, cat in enumerate(lib_arts_cats):
    x = data["Year"]
    y = data[cat]
    z = 100 - y
    axes[i].plot(x, y, label="Women")
    axes[i].plot(x, z, label="Men")
    axes[i].set_title(cat)
    axes[i].legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, len(other_cats), figsize=(25, 5))

for i, cat in enumerate(other_cats):
    x = data["Year"]
    y = data[cat]
    z = 100 - y
    axes[i].plot(x, y, label="Women")
    axes[i].plot(x, z, label="Men")
    axes[i].set_title(cat)
    axes[i].legend()
plt.show()
```

### üìà Observations

#### üî¨ STEM Fields:
- **Computer Science**: Saw a rise in female graduates until mid-1980s, then a decline.
- **Biology**: Women now dominate the field.
- **Engineering**: Still male-dominated, but women‚Äôs share has grown slightly.

#### üìö Liberal Arts:
- Most fields have consistently higher percentages of women, especially in **Foreign Languages** and **English**.

#### üè• Other Fields:
- **Health Professions** and **Education** are strongly female-dominated.
- **Business** shows a balanced gender trend in recent years.

---

### üí° Insights
- The gender gap in higher education has shifted significantly over time.

- Some traditionally male-dominated fields still show persistent gender imbalance.
